# Data Preparation

## Dataset Overview

### Data Source
Los Angeles restaurant data from public sources.

### Dataset Description

**Table: rest_data**

| Column | Description |
|--------|-------------|
| `id` |  Restaurant establishment unique id |
| `object_name` |  Restaurant establishment name |
| `address` | Full address of the establishment |
| `chain` | Whether the establishment is part of a chain (TRUE/FALSE) |
| `object_type` | Type of establishment (e.g., restaurant, cafe, bar) |
| `number` | Number of seats in the establishment |


## Objectives
The main objective of this notebook/steps is to verify data integrity and correct possible anomalies. Data will be loaded and some verifications will be performed to ensure the best quality for EDA.

1. **Load** the restaurant data from the CSV file
2. **Validate** data types for each column
3. **Check** for missing values and duplicates
4. **Process** the data if necessary
5. **Ensure** data quality for subsequent exploratory analysis (EDA)

```python
# Importing necessary libraries
import pandas as pd
```

```python
# Loading the dataset
df_restaurants = pd.read_csv("../data/rest_data_us_upd.csv")
```

```python
# Checking dataframe info
df_restaurants.info()
```

```python
# Checking df first 5 rows
df_restaurants.head(5)
```

- Dataframe have a total of 9651 rows;
- No null values in columns except for'chain' column that have 3 NaN; due to these NaN values variable type is not correct (should be 'boolean');
- Column 'object_type' can possible be a categorical variable; content need to be cheked;
- Other columns are already in correct type.

```python
# Drop rows with NaN values in 'chain' column
df_restaurants.dropna(subset=['chain'], inplace=True)
```

```python
# Converting column 'chain' to boolean type
df_restaurants['chain'] = df_restaurants['chain'].astype(bool)
```

```python
# Checking 'chain' distribution
df_restaurants['chain'].value_counts(normalize=True)
```

```python
# Checking duplicated rows in df
dup_rows = df_restaurants.duplicated().sum()
print(f'df_restaurants have {dup_rows} duplicated rows')
```

```python
# Checking id column integrity
unique_id = df_restaurants['id'].nunique()
print(f'id column have {unique_id} unique values')
```

```python
# Checking names unique values
df_restaurants['object_name'].nunique()
```

```python
# Checking establishments options
df_restaurants['object_type'].value_counts()
```

```python
# Converting 'object_type' to category
df_restaurants['object_type'] = df_restaurants['object_type'].astype(
    'category')
```

```python
# Checking establishments number of seats distribution
df_restaurants['number'].describe()
```

```python
# Checking duplicated values in 'address' column
dup_address = df_restaurants['address'].duplicated().sum()
print(f'df_restaurants have {dup_address} duplicated addresses')
```

```python
# Checking duplicated values in 'address' column
df_restaurants[df_restaurants['address'].duplicated(
    keep=False)].sort_values('address').head(10)
```

```python
# Final dataframe validation
df_restaurants.info()
```

### Observations
- 'id' column has no duplicated values, which confirms data integrity as this should be a unique identifier;
- 'object_name' has a total of 8669 unique names out of 9648 rows, meaning 979 names appear more than once — these could be chain establishments or independent restaurants that share the same name; further analysis needed in EDA;
- 'object_type' distribution shows most of the data refers to 'Restaurants' (7253), followed by 'Fast Food' (1066) and then 'Cafe' (435), 'Pizza' (319) and 'Bakery' (283);
- 'number' column has a correct range, between 1 and 229 seats; looking at the mean of 43.70 and median of 27 we have a right-skewed distribution, possibly due to some higher values (e.g., max is 229 seats);
- 'address' column has 1139 duplicated values; we see a street with 2 restaurants and '100 WORLD WAY' looks like a shopping center as we have several types of establishments.
